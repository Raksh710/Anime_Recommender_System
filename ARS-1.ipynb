# Calculating the number of tokens

```python
import tiktoken
```

```python
def num_tokesns_from_string(string, encoding_name):
    encoding = tiktoken.get_encoding(encoding_name)
    num_tokens = len(encoding.encode(string))
    return num_tokens
```

```python
num_tokesns_from_string(string = df_popular_anime['summary'][0],encoding_name = 'cl100k_base') # name of encoding used by Open AI cl100k_base
```

```python
len(df_popular_anime['summary'][0].split()) # number of words
```

```python
df_popular_anime['token_counts'] = df_popular_anime['summary'].apply(lambda x: num_tokesns_from_string(string=x,encoding_name="cl100k_base"))
```

```python
def unknown_changer(x):
    if x=='Unknown':
        return 0
    else:
        return np.float64(x)
```

```python
df_popular_anime['Score'] = df_popular_anime['Score'].apply(lambda x: unknown_changer(x))
```
